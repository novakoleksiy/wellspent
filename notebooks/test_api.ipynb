```python
import httpx

BASE = "http://localhost:8000"
client = httpx.Client(base_url=BASE)
TOKEN = None
```

## Health

```python
client.get("/health").json()
```

## Register

```python
r = client.post("/api/users/register", json={
    "email": "test@example.com",
    "password": "secret123",
    "full_name": "Test User",
})
print(r.status_code)
r.json()
```

## Login

```python
r = client.post("/api/users/login", json={
    "email": "test@example.com",
    "password": "secret123",
})
TOKEN = r.json()["access_token"]
client.headers.update({"Authorization": f"Bearer {TOKEN}"})
print("Logged in, token:", TOKEN[:30], "...")
```

## Get current user

```python
client.get("/api/users/me").json()
```

## Update preferences

```python
client.put("/api/users/me/preferences", json={
    "budget_tier": "mid",
    "travel_styles": ["adventure", "foodie"],
    "accommodation_types": ["hotel"],
    "pace": "moderate",
    "notes": "Love rooftop bars",
}).json()
```

## Get recommendations

```python
client.post("/api/trips/recommend", json={
    "destination": "Tokyo, Japan",
    "start_date": "2026-06-01",
    "end_date": "2026-06-05",
    "travelers": 2,
    "budget_max": 3000,
}).json()
```

## Create trip

```python
r = client.post("/api/trips/", json={
    "title": "Tokyo Adventure",
    "destination": "Tokyo, Japan",
    "description": "A 4-day trip to Tokyo",
})
TRIP_ID = r.json()["id"]
print("Created trip id:", TRIP_ID)
r.json()
```

## List trips

```python
client.get("/api/trips/").json()
```

## Get single trip

```python
client.get(f"/api/trips/{TRIP_ID}").json()
```

## Delete trip

```python
r = client.delete(f"/api/trips/{TRIP_ID}")
print(r.status_code)  # 204
```
