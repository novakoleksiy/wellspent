```python
import sqlite3
import json
import pandas as pd

DB_PATH = "../travel.db"
con = sqlite3.connect(DB_PATH)
con.row_factory = sqlite3.Row

def q(sql, params=()):
    rows = con.execute(sql, params).fetchall()
    if not rows:
        return pd.DataFrame()
    return pd.DataFrame([dict(r) for r in rows])
```

## Tables

```python
q("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name")
```

## Users

```python
q("SELECT id, email, full_name, preferences, created_at FROM users")
```

## Trips

```python
q("SELECT * FROM trips")
```

## Bookings

```python
q("SELECT * FROM bookings")
```

## Trips with user email

```python
q("""
SELECT t.id, u.email, t.title, t.destination, t.status, t.created_at
FROM trips t
JOIN users u ON u.id = t.user_id
ORDER BY t.id
""")
```

## Itinerary detail (pretty-print JSON)

```python
TRIP_ID = 1
row = con.execute("SELECT itinerary FROM trips WHERE id=?", (TRIP_ID,)).fetchone()
if row and row["itinerary"]:
    print(json.dumps(json.loads(row["itinerary"]), indent=2))
else:
    print("No itinerary found for trip", TRIP_ID)
```

## Row counts

```python
pd.DataFrame([
    {"table": t, "rows": con.execute(f"SELECT COUNT(*) FROM {t}").fetchone()[0]}
    for t in ["users", "trips", "bookings"]
])
```
